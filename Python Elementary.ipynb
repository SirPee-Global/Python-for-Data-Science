# Variable and Identifier

## Rules for Naming Variable
1. Variable name should be descriptive
2. Variable name cannot start with number, though number can be find anywher in it
3. Any special characters except underscore cannot be used in naming variable
4. Space cannot be used to join words in naming variables, used underscore instead


```python
first_name = "Promise"
last_name = "Chukwudi"
age = 12

print("my name is", first_name, last_name, "and iam", age, "years old")
```

## Arithmetic

```python
# addition (+)
a = 12
b = 10
c = a + b 
print (c)
```

```python
# multiplication (*)
a = 12
b = 10
c = a * b 
print (c)
```

```python
# divsion (/)
a = 12
b = 10
c = a / b 
print (c)
```

## Assignment Operator

```python
# equal to (=)
name = "Promise"
print(name)
```

```python
# increment (+=)
a = 10
a+=5 # increase the value of a by five
print(a)
```

```python
# decrement (-=)
a = 10
a-=3 # decrease the value of a by three
print(a)
```

## Comparison Operator 
is used to compare variables, and it return boolean result

```python
# equal equal to (==)
a = 4
b = 7
a == b # check if a is the same as b
```

```python
# greater than (>)
a = 4
b = 7
a > b # check if a is greater than b
```

```python
# greater than  or equal (>=)
a = 7
b = 7
a >= b # check if a is greater than or equal b
```

```python
# less than (<)
a = 4
b = 7
a < b # check if a is lesser than b
```

```python
# less than  or equal (<=)
a = 4
b = 7
a <= b # check if a is less than or equal b
```
